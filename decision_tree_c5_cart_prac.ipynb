```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv(r'IRIS.csv')
iris.head(60)
```

```python
# conver categorical data to numerical using label encodding: 
label_encoder = preprocessing.LabelEncoder()
iris['species'] = label_encoder.fit_transform(iris['species'])
iris.head()
```

```python
# split the data :
x = iris.iloc[:,0:4]
y = iris.iloc[:,4]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)
# x_train.shape
```

```python
# C5.0 :
model = DecisionTreeClassifier(criterion='entropy', max_depth=3)
model.fit(x_train, y_train)
tree.plot_tree(model)
```

```python
fn = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
cn = ["setosa", "versicolor", "virginica"]
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(4,4), dpi=300)
tree.plot_tree(model, feature_names=fn, class_names = cn, filled=True)
```

```python
pred = model.predict(x_test)
pd.Series(pred).value_counts()
```

```python
# to check accuracy score :
import numpy as np
np.mean(pred==y_test)           # 0.96667
pd.crosstab(y_test, pred)
```

```python
# CART (classification and regression tree) :
# for classification :
model_gini = DecisionTreeClassifier(criterion='gini', max_depth=3)
model_gini.fit(x_train, y_train)
gini_pred = model_gini.predict(x_test)
pd.Series(gini_pred).value_counts()
np.mean(y_test==gini_pred)     # 0.966667
```

```python
# DecisionTree for Regression :
array = iris.values
# here we cant take categorical columns (species) so we try to predict petal_width :
x_new = array[:,0:3]
y_new = array[:,3]
x_new_train, x_new_test, y_new_train, y_new_test = train_test_split(x_new, y_new, test_size=0.3, random_state=40)
# x_new_train.shape                          # (105, 3)
```

```python
# make model for regression and fit :
reg_model = DecisionTreeRegressor()
reg_model.fit(x_new_train, y_new_train)
reg_model.score(x_new_train, y_new_train)             # 0.999665
reg_pred = reg_model.predict(x_new_test)
pd.crosstab(reg_pred, y_new_test)
np.mean(reg_pred==y_new_test)                    # 0.222
```
